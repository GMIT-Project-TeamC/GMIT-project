```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import matplotlib as mpl
```

```python
frameA = pd.read_csv("stock-prices.csv", dtype=str, delimiter=",", skiprows = None, parse_dates=['Date'])
```

```python
frameB = pd.read_csv("tweets.csv", dtype=str, delimiter=",", skiprows = None, parse_dates=['Date'])
```

```python
df_a = pd.DataFrame(data=frameA)
df_b = pd.DataFrame(data=frameB)
```

```python
merged = df_a.merge(df_b, on='Date')
```

```python
merged.to_csv('out.csv')
```

```python
stock_filepath = "out.csv" 

stock_temp = pd.read_csv(stock_filepath, header = 0, index_col = 2,
                         parse_dates=True, infer_datetime_format=True)


ax = stock_temp.plot(figsize=(20,10))
```
