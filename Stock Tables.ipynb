```python
import pandas as pd
import numpy as np

stock_filepath = "stock-prices.csv" 
tweet_filepath = "tweets.csv" 

stock_temp = pd.read_csv(stock_filepath, header = 0, index_col = 0,
                         parse_dates=True, infer_datetime_format=True)
tweet_temp = pd.read_csv(tweet_filepath, header = 0, index_col = 0,
                         parse_dates=True, infer_datetime_format=True)

#plot both overlayed
ax = tweet_temp.plot(figsize=(20,10))
stock_temp.plot(figsize=(20,10))
```
